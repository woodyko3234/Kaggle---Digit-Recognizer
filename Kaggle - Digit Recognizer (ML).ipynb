### 1. Read the dataset and check if anything missing

```python
#import necessary modules to manipulate the datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read training and test sets
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train.describe()
```

```python
test.describe()
```

```python
#find nan values
print(np.sum(np.sum(train.isnull(), axis = 1)))
print(np.sum(np.sum(test.isnull(), axis = 1)))
```

Every sample is a 28*28 img

The first column "label" ranges from 0 to 9, which also should be the output value

No missing/NaN values in the datasets

```python
#set features and labels
features = train.iloc[:, 1:]
label = train.iloc[:, 0]
```

### 2. Try sklearn / machine learning

```python
#import ML modules
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
#logistic regression
clf = GaussianNB()
clf.fit(features, label)
```

```python
clf.score(features, label)
```

All other ML algorithms take a lot of time training.

And Gaussian got a horrible score.

There's no way implementing them without normalization and grouping

```python
#import preprocessing module
from sklearn.preprocessing import MinMaxScaler
normalized_features = MinMaxScaler().fit_transform(features.loc[:, features.columns])
normalized_test = MinMaxScaler().fit_transform(test.loc[:, test.columns])
```

```python
normalized_features.shape
```

```python
clf.fit(normalized_features, label)
clf.score(normalized_features, label)
```

```python
import time
```

```python
# Try LogisticRegression
print(time.asctime())
clf = LogisticRegression()
clf.fit(normalized_features, label)
print(clf.score(normalized_features, label))
print(time.asctime())
```

```python
#pretty good score
#first try on test and upload to Kaggle
#normalized_test = MinMaxScaler().fit_transform(test.loc[:, test.columns])
prediction = pd.DataFrame(clf.predict(normalized_test), columns = ['Label']).reset_index()
```

```python
prediction.rename(columns = {'index': 'ImageId'}, inplace = True)
prediction.iloc[:, 0] += 1
prediction
```

```python
prediction.to_csv('logisticML.csv', index = False)
#accuracy = 0.91642
```

```python
#how about knn
#import time
print(time.asctime())
clf = KNeighborsClassifier()
clf.fit(normalized_features, label)
#print(clf.score(normalized_features, label)) #0.979142857143
print(time.asctime())
```

```python
#even better!
#normalized_test = MinMaxScaler().fit_transform(test.loc[:, test.columns])
prediction = pd.DataFrame(clf.predict(normalized_test), columns = ['Label']).reset_index()
prediction.rename(columns = {'index': 'ImageId'}, inplace = True)
prediction.iloc[:, 0] += 1
prediction.to_csv('knnML.csv', index = False)
#0.96800
#but took a lot of time to predict!!!
```

```python
#decision tree
print(time.asctime())
clf = DecisionTreeClassifier()
clf.fit(normalized_features, label)
print(clf.score(normalized_features, label))
print(time.asctime())
```

```python
print(time.asctime())
prediction = pd.DataFrame(clf.predict(normalized_test), columns = ['Label']).reset_index()
prediction.rename(columns = {'index': 'ImageId'}, inplace = True)
prediction.iloc[:, 0] += 1
prediction.to_csv('dtML.csv', index = False)
print(time.asctime())
#0.85485 sucks
```
